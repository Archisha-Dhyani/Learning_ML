```python
df_setosa= df.loc[df['Species']=='setosa']
df_virginica=df.loc[df['Species']=='virginica']
df_versicolor=df.loc[df['Species']=='versicolor']
print(df_setosa.shape,df_virginica.shape,df_versicolor.shape)
```

# Uni-Variate- Analysis

```python
# zeros_like() function in python returns an array of zeros with teh same shape as width in this case
plt.plot(df_setosa['Sepal.Width'],np.zeros_like(df_setosa['Sepal.Width']),'o')
plt.plot(df_virginica['Sepal.Width'],np.zeros_like(df_virginica['Sepal.Width']),'o')
plt.plot(df_versicolor['Sepal.Width'],np.zeros_like(df_versicolor['Sepal.Width']),'o')
plt.xlabel('SepalWidth')
plt.legend(['setosa','virginica','versicolor'],loc=1)
plt.show()
```

```python
plt.plot(df_setosa['Sepal.Length'],np.zeros_like(df_setosa['Sepal.Length']),'o')
plt.plot(df_virginica['Sepal.Length'],np.zeros_like(df_virginica['Sepal.Length']),'o')
plt.plot(df_versicolor['Sepal.Length'],np.zeros_like(df_versicolor['Sepal.Length']),'o')
plt.xlabel('SepalLength')
plt.legend(['setosa','virginica','versicolor'],loc=1)
plt.show()
```

```python
species_names = df['Species'].unique()
print(species_names)
```

# Bi-variate- Analysis

```python
#print scatter graph between two attribotes for each species one by one 
for Species in species_names:
    species_data = df[df['Species'] == Species]
    plt.scatter(species_data['Petal.Length'],species_data['Sepal.Width'], label=Species)
    
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot - Petal Length vs Sepal Width')
plt.legend()
plt.show()
```
